```python
df = pd.read_csv("df_social.csv")
df
```

```python
years = np.unique([x.split("_")[1] for x in df.columns[1:]])
```

```python
df_base = pd.DataFrame(df.iloc[:,0])
```

```python
for mode in ["IN", "OUT"]:
    for sex in ["F", "M"]:
        for year in years:
            my_regex = mode + "_" + year + "_" + sex + ".*"
            temp = df.filter(regex = my_regex)
            temp = temp.mean(axis = 1)
            df_base[my_regex] = temp
```

```python
pd.get_option("display.max_columns", None) 
df_base = df_base.rename(columns = {"Unnamed: 0" : "CODE"})
```

```python
df_base
```

```python
df_base.columns = [x.replace(".*", "") for x in df_base.columns]
```

```python
df_base
```

```python
df_base.to_csv("df_subjective_norm_temp.csv", index = False)
```
